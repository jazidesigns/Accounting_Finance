**1. Calculate Gross Profit Margin**

```python
gross_profit = 200000
revenue = 800000

margin = (gross_profit / revenue) * 100
print(f"Gross Profit Margin = {margin:.2f}%")
```

**2. Calculate Net Profit Margin**

```python
net_profit = 50000
revenue = 150000

margin = (net_profit / revenue) * 100
print(f"Net Profit Margin = {margin:.2f}%")
```

**3. Calculate Return On Assets (ROA)**

```python
net_income = 80000
total_assets = 500000

roa = (net_income / total_assets) * 100
print(f"Return On Assets = {roa:.2f}%")
```

**4. Calculate Return On Equity (ROE)**

```python
net_income = 60000
shareholders_equity = 400000

roe = (net_income / shareholders_equity) * 100
print(f"Return On Equity = {roe:.2f}%")
```

**5. Calculate Earnings Per Share (EPS)**

```python
net_income = 150000
dividends = 30000
num_shares = 10000

eps = (net_income - dividends) / num_shares
print(f"Earnings Per Share = ${eps:.2f}")
```

**6. Calculate Price To Earnings Ratio**

```python
stock_price = 50
eps = 12
pa_ratio = stock_price / eps
print(f"Price To Earnings Ratio = ${pa_ratio:.2f}")
```

**7. Calculate Depreciation Expense (Straight-Line Method)**

```python
initial_cost = 50000
salvage_value = 10000
useful_life = 5

depreciation_expense = (initial_cost - salvage_value) / useful_life
print(f"Depreciation Expense = ${depreciation_expense} per year")
```

**8. Calculate Working Capital**

```python
current_assets = 60000
current_liabilities = 40000

working_capital = current_assets - current_liabilities
print("Working Capital = $" + str(working_capital))
```

**9. Calculate Current Ratio**

```python
assets = 80000
liabilities = 50000

current_ratio = (assets / liabilities)
print("Current Ratio =", current_ratio)
```

**10. Calculate Quick Ratio (Acid-Test Ratio)**

```python
current_assets = 80000
inventory = 20000
current_liabilities = 50000

quick_ratio = (current_assets - inventory) / current_liabilities
print("Quick Ratio =", quick_ratio)
```

**11. Calculate Debt to Equity Ratio**

```python
total_debt = 50000
shareholders_equity = 70000

debt_to_equity_ratio = total_debt / shareholders_equity
print("Debt To Equity Ratio:", round(debt_to_equity_ratio, 2))
```

**12. Calculate Dividend Yield:**

```python
dividends_per_share = 2.5
stock_price = 50

dividend_yield = (dividends_per_share / stock_price) * 100
print(f"Dividend Yield = {dividend_yield:.2f}%")
```

**13. Price To Book (P/B) Ratio**

```python
stock_price = 60
book_value_per_share = 40

pb_ratio = stock_price / book_value_per_share
print("Price To Book Ratio = " + str(pb_ratio))
```

**14. Balanse Sheet**

```python
import pandas as pd

# Create a sample balance sheet dataset
data = {
    'Category': ['Assets', 'Assets', 'Assets', 'Liabilities', 'Liabilities', 'Equity'],
    'Account': ['Cash', 'Inventory', 'Equipment', 'Accounts Payable', 'Loan Payable', 'Owner\'s Equity'],
    'Amount': [50000, 30000, 60000, 25000, 40000, None]
}

# Create a DataFrame from the dataset
balance_sheet_df = pd.DataFrame(data)
balance_sheet_df 
```

```python
assets = balance_sheet_df[balance_sheet_df["Category"] == "Assets"]
total_assets = assets["Amount"].sum()
total_assets
```

```python
liabilities =  balance_sheet_df[balance_sheet_df["Category"] == "Liabilities"]
total_liabilities = liabilities["Amount"].sum()
total_liabilities
```

```python
total_equity = total_assets - total_liabilities
total_equity
```

```python
is_balanced = total_assets == total_liabilities + total_equity
print("Is Balance Sheet Balanced?", is_balanced)
```

**15. Income Statement**

```python
data = {
    'Category': ['Revenue', 'Expenses', 'Expenses', 'Expenses'],
    'Account': ['Sales', 'Salaries', 'Rent', 'Supplies'],
    'Amount': [100000, 30000, 15000, 5000]
}

# Create a DataFrame from the dataset
income_statement_df = pd.DataFrame(data)

income_statement_df
```

```python
revenue = income_statement_df[income_statement_df["Category"] == "Revenue"]
total_revenue = revenue["Amount"].sum()
total_revenue
```

```python
expense = income_statement_df[income_statement_df["Category"] == "Expenses"]
total_expense = expense["Amount"].sum()
total_expense
```

```python
net_income = total_revenue - total_expense
net_income
```
