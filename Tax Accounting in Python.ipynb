**1. Income Tax Calculation**

```python
def income_tax(income):
    if income <= 10000:
        tax = income * 0.10
    elif income <= 50000:
        tax = 10000 * 0.10 + (income - 10000) * 0.20
    else:
        tax = 10000 * 0.10 + 40000 * 0.20 + (income - 50000) * 0.30
    return tax
```

```python
income = 45000
tax = income_tax(income)
print(f"Income Tax for ${income} = ${tax:.2f}")
```

**2. Sales Tax Calculation**

```python
purchase_amount = 100
item_type = "food"
```

```python
def sales_tax(item_type, purchase_amount):
    tax_rate = {
        "clothing": 0.05,
        "electronics": 0.10,
        "food": 0.02
    }
    tax_rate = tax_rate.get(item_type, 0.0)
    sales_tax = purchase_amount * tax_rate
    return sales_tax
```

```python
sales_tax = sales_tax(item_type, purchase_amount)
print("Sales Tax = $" + str(sales_tax))
```

**3. Depreciation Calculation**

```python
cost = 10000
salvage_value = 2000
useful_life = 5

depreciation = (cost - salvage_value) / useful_life
print("Depreciation Per Year = $" + str(depreciation))
```

**4. Tax Withholding Calculator**

```python
def tax_withholding(income, dependents, tax_rate, deduction_per_dependent):
    withholding_tax = (income - dependents * deduction_per_dependent) * tax_rate
    return withholding_tax
```

```python
income = 50000
dependents = 2
tax_rate = 0.20
deduction_per_dependent = 1000
```

```python
withholding_tax = tax_withholding(income, dependents, tax_rate, deduction_per_dependent)
```

```python
print(f"Withholding Tax for ${income} with {dependents} dependents: ${withholding_tax}")
```

```python
def calculate_property_tax(property_value, tax_rate):
    property_tax = property_value * (tax_rate)
    return property_tax

property_value = 250000
tax_rate = 0.012
property_tax = calculate_property_tax(property_value, tax_rate)
print(f"Property Tax for a property valued at ${property_value}: ${property_tax:.2f}")
```

**5. Property Tax Assessment**

```python
def calculate_property_tax(property_value, tax_rate):
    property_tax = property_value * (tax_rate / 100)
    return property_tax
```

```python
property_value = 250000
tax_rate = 1.2
```

```python
property_tax = calculate_property_tax(property_value, tax_rate)
```

```python
print(f"Property Tax for a property valued at ${property_value} = ${property_tax}")
```
